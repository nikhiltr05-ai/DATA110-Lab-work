```python
stats = df.describe()
stats
```

```python
df.hist(figsize=(16,12))
plt.tight_layout()
plt.show()
```

```python
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    plt.figure()
    sns.histplot(df[col].dropna(), kde=True)
    plt.show()
```

```python
html = f"""
<h1>Facebook Friends Data Analysis</h1>
<h2>Descriptive Statistics</h2>
{stats.to_html()}
<h2>Missing Values</h2>
{df.isnull().sum().to_frame("Missing Values").to_html()}
"""
open("Facebook_Data_Analysis.html","w").write(html)
"HTML Created"
```
