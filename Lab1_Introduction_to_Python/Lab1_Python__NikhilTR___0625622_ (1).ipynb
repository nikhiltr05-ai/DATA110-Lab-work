container types in python
The four core built-in container types are:
list: A mutable, ordered sequence of elements. Lists are indexed by integers and can hold items of different types. They are defined using square brackets (e.g., [1, "a", 2.5]).
tuple: An immutable, ordered sequence of elements. Tuples are similar to lists but cannot be changed after creation. They are defined using parentheses (e.g., (1, "a", 2.5)).
dict (Dictionary): A mutable mapping type that stores data as unordered (ordered since Python 3.6) key-value pairs. Data is accessed using unique keys, not integer indices. They are defined using curly braces (e.g., {"name": "John", "age": 30}).
set: A mutable, unordered collection of unique elements. Sets are useful for operations like union and intersection and for quickly checking membership. They are defined using curly braces (e.g., {1, 2, 3}) or the set() constructor.

Problem 1: Confusion Matrix Metrics (Python as a Calculator)
A binary classification model predicts whether rainfall will cause Flood (Yes/No).
Given:
	True Positives (TP) = 50
	False Positives (FP) = 8
	True Negatives (TN) = 32
	False Negatives (FN) = 10
Tasks
	Calculate total predictions
	Compute:
	Accuracy
	Precision
	Recall
	F1 Score

Formulae
"Accuracy"=(TP+TN)/(TP+FP+TN+FN)
"Precision"=TP/(TP+FP)
"Recall"=TP/(TP+FN)
"F1"=2PR/(P+R)


```python
# Given values
TP = 50
FP = 8
TN = 32
FN = 10

# Total predictions
total = TP + FP + TN + FN

# Metrics
accuracy = (TP + TN) / total
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * precision * recall / (precision + recall)

# Output
print("Total Predictions:", total)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```

Problem 2: R² (Coefficient of Determination)
In a regression model predicting river discharge, the following are known:
	Sum of Squares due to Regression (SSR) = 420
	Sum of Squares due to Error (SSE) = 180
Tasks
	Compute Total Sum of Squares (SST)
	Calculate R²
	Interpret whether the model explains more than 70% variance

Formula
SST=SSR+SSE
R^2=SSR/SST

```python
# Given values
SSR = 420
SSE = 180

# Total Sum of Squares
SST = SSR + SSE

# R-squared
R_squared = SSR / SST

# Output
print("SST:", SST)
print("R^2:", R_squared)
```

Problem 3: Mean Squared Error (MSE) Using Python as Calculator
A model predicts temperature values.
The squared errors for 5 observations are:
(2ⓜ,4ⓜ,1ⓜ,9ⓜ,4)

Tasks
	Store squared errors in Python
	Calculate Mean Squared Error (MSE)
	Print the result

Formula
MSE=(∑e^2)/n


```python
# Squared errors
squared_errors = [2, 4, 1, 9, 4]

# Number of observations
n = len(squared_errors)

# Mean Squared Error
mse = sum(squared_errors) / n

# Output
print("Mean Squared Error (MSE):", mse)
```

Problem 4: Cook’s Distance (Influential Observation)
For a regression model:
	Squared residual (eᵢ²) = 9
	Mean Squared Error (MSE) = 3
	Number of predictors (p) = 3
	Leverage value (hᵢᵢ) = 0.2
	Sample size (n) = 25
Tasks
	Compute Cook’s Distance
	Compare with threshold 4/n
	Decide whether the observation is influential

Formula
D_i=(e_i^2)/(p×MSE)×h_ii/(1ⓜ-h_ii )^2 ┤


```python
# Given values
e_sq = 9
MSE = 3
p = 3
h_ii = 0.2
n = 25

# Cook's Distance
D_i = (e_sq / (p * MSE)) * (h_ii / (1 - h_ii)**2)

# Threshold
threshold = 4 / n

# Output
print("Cook's Distance:", D_i)
print("Threshold (4/n):", threshold)
print("Influential Observation:", D_i > threshold)
```
